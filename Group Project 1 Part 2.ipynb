   Exercise is an imperative part of health human, granting numerous benefits to the human body and mind. This mentality is becoming increasingly popular with the presence of obesity and stress within society. People are beginning to include exercise as a part of their daily routine by making use of their free time in the morning or night. Thus, in order to help individuals track their workouts and the effectiveness of it, we borrowed the idea of utilizing a fitness tracker similar to the Fitbit. Our proposition is a fitness tracker that serves to show users which type of exercise routine is the most effective in burning calories. The fitness tracker is capable of tracking on ground activities, like walking, running, speed walking, and biking, and underwater activities, like swimming. More routine-based and intensive workouts are also trackable with the inclusion of sets and repetitions completed by the user. Yet, unlike other Fitbits that are able to keep track of sleep patterns, the tracker is solely fitness based and meant to be used for exercise only; however, it can also serve as a watch with the time display installed in the bracelet. 
   
   Using the fitness tracker is relatively easy. Each workout will be recorded once the user presses the start button. Once the start button is pressed, the user may begin his workout and the tracker will provide a display of the ongoing workout duration and the amount of calories burned. If the user wishes to stop the workout or to begin a new set, he may press the stop button and end his activities or begin a new set. However, if the user wishes to do a repetition, then he may press the restart button. Once the user is finished with his workout, he will be provided with a report, called the workout log, of his activities conducted. This workout log serves as the end-user data by granting individuals information on the type of activity performed, number of sets and repetition conducted, duration of workout in minutes, number of calories burned per minute, and heart rate in beats per minute. 
Overall, the objective of the fitness tracker is to help users detect which activities are the most effective in burning calories. In order to do this, we will utilize the tables, workout log and calories.

Tables:

1)The first table is user which provides data about the user such as first name, last name, age, gender, and weight. The entity of this table is the user and the attributes include first name, last name, age, gender, and weight of the user. 

2)The second table is activity which includes information on the type of exercises conducted by the user. The entity of this table is activity and the attributes are walking, running, speed walking, biking, and swimming. 

3)The third table is workout which gives information on the number of sets and repetitions completed by the user. The entity is workout and the attributes are number of sets and repetitions.

4)The fourth table is time which provides information on the duration of the workout conducted by the user. The entity is time and the attributes are hours, minutes, seconds, and milliseconds. 

5)The fifth table is calories which provides data on the amount of calories burned during exercising. The entity is calories and the attributes are calories burned per hour, calories burned per minute, and calories burned per second. 

6)The sixth table is heart rate which includes information on the user’s heart rate during a workout or a normal basis. The entity is heart rate and the attributes are beats per minute and beats per second.

7)The seventh table is the workout log, which is also known as the end-user data, that provides information on a summary of the user’s workout. The entity is workout log and the attributes are type of activity, number of sets, number of repetitions, duration of workout in minutes, number of calories burned per minute, and heart rate in beats per minute. By using the workout log, users will be able to determine which workout or activity is the most effective in burning calories.

Constraints:

The tracker is limited to only one user and may only track the calories burned by doing the activities (walking, speed walking, running, biking, and swimming) listed above
