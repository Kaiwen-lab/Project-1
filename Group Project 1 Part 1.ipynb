Idea 2: Creamery Inventory 

   Penn State’s Berkey Creamery is a historic landmark for students and the public since 1965. Enjoying ice cream from the Creamery is a special part of the Penn State experience; the methods developed by the food and dairy scientists have made the Creamery a world authority on dairy manufacturing and ice cream production. But with over 100 ice cream flavors, 10 frozen yogurts, and 6 sherberts------ which is the favorite among the people and generates the most revenue? Our goal is to create a database for the Creamery to collect data that includes store employees, order type, flavors, additional products, and a table to collect customer purchases to decide which flavors are the most popular. As an existing business we would be able to retrieve data and relay this information back to the Creamery. At any given time there are only about 20 flavors offered at the store front. With the information from the database we would be able to display flavors that are more popular to generate the  most revenue and draw more customers. The customer transaction is quite simple. When you step foot through the doors of the Creamery you can look for additional products in front of the counter. When you finally reach the register, you give the employee your order and any other items you are purchasing and give a form of payment. This may include lion cash, meal point, credit/debit cards, or cash.
    
Tables:

1)Store Employees: this will include data of employees working and their position for their shift. This means cashier, ice cream scooper, milkshake, stock, etc.

Entities/Attributes: employee name- first name, last name; store ID- number combination; address- street, city, state, zip code; phone- area code, phone number; position- type of position

2)Order Type: Pick cone, cup, or milkshake? What size? Then, what is the corresponding price for the combination.

Entities/Attributes: cone- waffle, wafer, price; cup- regular, large, price; milkshake- price

3)Flavors: List the type of flavors that are offered at the creamery with a brief description of the flavor and ingredients.

Entities/Attributes: flavors- description of the flavor, ingredients

4)Additional Products: There are additional products offered to be purchased at the Creamery, and this table would include all non-ice cream products sold in front of the counter. 

Entities/Attributes: name; cost; manufacturer- address, name; expiration date- date arrived, date best used by

5)Customer Purchases: This table we would use to collect customer purchases and keep track of the quantity of each flavor sold. This is our “end-user” data. Collect data from the customer and keep track of their purchases.

Entities/Attributes: customer name- first name, last name; address- street, city, state, zip code; phone- area code, number;  flavor sold- quantity of each flavor sold; transaction type- lion cash, meal points, credit/debit card, cash

Idea 3: Football Player’s data 

Goal: Track the stats of offensive football players to help analyze amount of pay they should receive
Our idea is to analyze the performance of a football players, who play on the offensive side of the football, through basic stats, to help teams manage the amount of money they should spend on a player. Although a team already has access to certain stats and data, what makes our idea unique is that it provides teams with an easier way of analyzing players stats based on trends over the past couple of seasons. It enables them to have access to a more organized stats format, making it easier to calculate the amount of money they should be giving to football players. This is an already existing business in that sports websites track these stats, but they don’t organize it very well, so we are trying to organize the data in a way that makes sense, so that it’s easier for football teams to analyze performance. For example, a team can easily compare the amount of money two different players, which is a potential interaction you may see with our product, in this case say Player A is getting paid higher than Player B, but has a weaker performance than Player B, it would help the football team understand that Player B is playing better than Player A, so during Player B’s contract year, the team can pay Player B the amount that he truly deserves based on his performance.

Tables:

1)Player Name – Name of the player
Entities/Attributes: First Name, Last Name

2)Player Position – Position the player plays (Ex: Quarterback, running back, etc.)

Entities/Attributes: Primary position, secondary position (if applicable)

3)Experience – number of years the player has played football

Entities/Attributes: Total experience, number of teams

4)Yard per season – number of yards in a season, respective to the player’s position (Ex: Quarterback – passing yards, Running Back – scrimmage yards, etc.)

Entities/Attributes: Total yards in career, individual season performance, strength of teams played in each season, averages, total games played

5)Touchdowns per season – number of total touchdowns in a season

Entities/Attributes: Total touchdowns in career, individual season performance, strength of teams played in each season, averages, total games played

6)Turnovers per season – amount of turnovers (interceptions and fumbles) in a season

Entities/Attributes: Total touchdowns in career, individual season performance, strength of teams played in each season, averages, total games played

7)Season Awards – lists awards won by the player in a season

Entities/Attributes: Award type, total awards, first place votes received

8)Amount paid – total amount player gets paid in the end result from performance

Our one dynamic data table is the amount the player gets paid because this is the end result the user (a team) must examine based on a players stats from current and previous seasons, to determine the salary needed for a player.
