As we are finding health and fitness data, we need to know the users that we get data from. Thus, an entity of user is created with attributes such as name, age, sex, height, and weight under it. By using the fitness tracker, the user is able to choose which activities to perform. Therefore, an entity of activities will be created along with an attribute of activity type. Also, two foreign keys are included which are user ID and workout log ID. These two foreign keys provide link to the user and workout log entity. The workout table is created as a means of end-user data, in order to allow the user access to the information from his or her workout. Attributes, like steps, elevation, milage, and pace, are included under this entity along with a few foreign keys, such as time duration, heart rate ID, and calories count. Furthermore, these three foreign keys have their own separate tables with their own unique attributes to provide information about the users workout. 

Along with creating entities, we must also define the relationships between each entity in our ER diagram. First off, a user can choose either zero activity or as many activities as he or she desires to, so this leads to a zero-to-many relationship. Secondly, a specific activity can be chosen by one and one user only so this is a one-to-one relationship. In addition, a specific activity will conduct a specific workout log and on the other hand, a specific workout log will connect to one activity. Both sides create a one-to-one relationship. Lastly, there are three sub-categories listed as foreign keys, time duration, heart rate ID, and calories count, under the workout log entity. Each of the three foreign keys produce a one-to-one relationship to the workout log since one specific workout will only have one specific time duration, heart rate, and calories count.
